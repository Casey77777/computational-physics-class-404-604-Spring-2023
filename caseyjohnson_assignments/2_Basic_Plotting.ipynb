Use sympy to integrate $f(x)$ with n=11 to derive the analytical expression after the integration ($g(x)$). And plot m(x)=g(x)-g(x=-$\pi$) from -$\pi$ to $\pi$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x,n):
    res = 0
    for k in range(1,n+2,2):
        res += np.sin(k*x)/k
    return np.pi/2 + 2*res
```

```python
xmin = -np.pi
xmax = np.pi

xs = np.linspace(xmin, xmax, 1000)

fig, axs = plt.subplots(1, figsize=(8,6))

axs.plot(xs, f(xs,11), alpha=0.3)
axs.plot(xs, f(xs,21), alpha=0.5)
axs.plot(xs, f(xs,31), alpha=1.0)
axs.set_xlim(xmin,xmax)
plt.grid(True)
```
